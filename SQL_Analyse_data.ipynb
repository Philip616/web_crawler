```python
def analyse_create(company_order):
    cursor = db.cursor()
    name = ""
    for i in range(len(company_order)-1):
        name = name + company_order.loc[i][0] + ' CHAR(10),'
    name = name + company_order.loc[len(company_order)-1][0] + " CHAR(10)"
    
    sql = "CREATE TABLE IF NOT EXISTS test ( Date DATE PRIMARY KEY, " + name + ")"
    cursor.execute(sql)
    
    cursor.close()
```

```python
def analyse_insert(company_order,data):
    cursor = db.cursor()
    
    name=""
    for i in range(len(company_order)-1):
        name = name + company_order.loc[i][0] + ','
    name = name + company_order.loc[len(company_order)-1][0]
    
    
    for i in range(len(data)):
        
        insert_data = ""
        insert_data = insert_data + "'" +data.loc[i][0] + "',"
        for j in range(1,len(company_order)):
            insert_data = insert_data + str(data.loc[i][j]) + ','
        insert_data = insert_data + str(data.loc[i][len(company_order)])

        sql = "INSERT INTO test (Date," + name + ") VALUES (" + insert_data + ")"
        
        print(sql)
        cursor.execute(sql)
        db.commit()
    cursor.close()
```
