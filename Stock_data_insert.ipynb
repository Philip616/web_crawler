```python
cursor = db.cursor()

sql = "show tables"
cursor.execute(sql)
sql_data = cursor.fetchall()
all_stock = pd.DataFrame(list(sql_data))

Date = pd.read_csv('../股票/国泰安Date.csv',engine="python")
Date['0'] = pd.to_datetime(Date['0'])

column = ['Date']
for i in range(len(all_stock)):
    print(all_stock.loc[i][0])
    
    column.append(all_stock.loc[i][0])
    sql = "select Date,Close from `" + all_stock.loc[i][0] + "`"
    cursor.execute(sql)
    sql_data = cursor.fetchall()
    Data = pd.DataFrame(list(sql_data),columns=['Date','Close'])
    
    Data['Date'] = pd.to_datetime(Data['Date'])
    
    Date = pd.merge(Date,Data,left_on=Date['0'],right_on=Data['Date'],how='left')
    del Date['Date']
    
Date.to_csv('../股票/Close.csv',index=None,header=column)
```
